```python
from sklearn.impute import KNNImputer
cont_cols = df.select_dtypes(include=['number']).columns
data2 = df[cont_cols] 
knn_imp = KNNImputer(n_neighbors=3)
data2 = pd.DataFrame(knn_imp.fit_transform(data2), columns=cont_cols)
print(data2.isnull().sum())
```

```python
X = data2.iloc[:,:-1]
X.head(2)
```

```python
y = data2.G3_x
y.head(2)
```

```python
y = data2.G3_y
y.head(2)
```

```python
X
```

```python
from sklearn.preprocessing import MinMaxScaler
sc  = MinMaxScaler()
final_X = pd.DataFrame(sc.fit_transform(X),columns=X.columns)
final_X
```

```python
final_X.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(final_X,y, test_size=0.2,random_state=42)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train , y_train)
```

```python
y_hat = model.predict(X_test)
y_hat
```

```python
plt.plot(y_hat,color='r')
plt.plot(np.array(y_test),color='g')
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mean_squared_error(y_test,y_hat)
```

```python
mean_absolute_error(y_test,y_hat)
```
