## Use KNNImputer (handling missing values in datasets by leveraging the K-Nearest Neighbors (KNN) algorithm.)

```python
from sklearn.impute import KNNImputer
cont_cols = df.select_dtypes(include=['number']).columns
data2 = df[cont_cols] 
knn_imp = KNNImputer(n_neighbors=3)
data2 = pd.DataFrame(knn_imp.fit_transform(data2), columns=cont_cols)
print(data2.isnull().sum())
```

```python
X = data2.iloc[:,:-1]
X.head(2)
```

```python
y = data2.G3_y
y.head(2)
```

```python
X.head()
```

## The MinMaxScaler from sklearn.preprocessing is used for feature scaling,

## The joblib library in Python is primarily used for efficient serialization, deserialization, and parallel computing.

```python
from sklearn.preprocessing import MinMaxScaler
import joblib
sc  = MinMaxScaler()
final_X = pd.DataFrame(sc.fit_transform(X),columns=X.columns)
final_X
```

```python
final_X.shape
```

```python
joblib.dump(sc,'scaler.pkl')
```

## The train_test_split function from sklearn.model_selection is used to split a dataset into training and testing sets

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(final_X,y, test_size=0.2,random_state=42)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

## The LinearRegression class from sklearn.linear_model is used to implement linear regression,

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train , y_train)
```

# represents the predicted values of the target variable

```python
y_hat = model.predict(X_test)
y_hat
```

# The graph you provided is an Elbow Curve, commonly used in K-Means Clustering to determine the optimal number of clusters (K).

```python
wcss = []
for i in range(2,20):
    dummy = KMeans(n_clusters=i, random_state=42)
    dummy.fit(final_X)
    wcss.append(dummy.inertia_)
```

```python
sns.lineplot(y=wcss, x=range(2,20))
plt.xticks(range(2,20))
plt.grid(axis='x')
plt.show()
```

```python
cluster_model = KMeans(n_clusters=3, random_state=42)
cluster_model.fit(final_X)
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(final_X, cluster_model.labels_)
```
